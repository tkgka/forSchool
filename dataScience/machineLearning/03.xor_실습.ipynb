```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#데이터 생성
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])

#2차원, 데이터개수 4개
X
#1차원, 데이터개수 4개
y
```

```python
#신경망 모델 생성
model = Sequential()
#은닉층
#units = 신경망 개수, dim = 입력이 들어오는 차원의 개수, x의 열의 개수
#활성화 함수 : relu, sigmoid, tanh
model.add(Dense(units=32, input_dim=2, activation="sigmoid"))
#출력층 2진분류는 출력층이 한개
model.add(Dense(units=1, activation="sigmoid"))#이진분류의 출력층 활성화 함수는 sigmoid
```

```python
model.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics=["accuracy"])
```

```python
model.summary()
```

```python
#error(loss) 와 optimizer 작성
model.compile(loss = "binary_crossentropy",#이진분류는 binary_crossentropy
             optimizer="adam",#RMSPorp
             metrics=["accuracy"])#평가
```

```python
#X: 문제 , y: 답 , epoch: 학습횟수 , verbose 출력창 제어(0:숨김, 1:세밀하게 , 2: 간단하게)
model.fit(X,y,epochs=10000, verbose=2)
```

```python
evalu = model.evaluate(X,y, verbose=0)
print("loss: ",evalu[0])
print("accuracy: ",evalu[1])
```

```python
#예측
model.predict(X)
#0.5보다 작은 경우 0으로 분류, 0.5보다 크면 1로 분류
```

```python
#에측
model.predict_classes(X)
```

```python
#가중치 확인
model.get_weights()
```

```python
#다중분류
x = np.array([[0,0],[0,1],[1,0],[1,1]])
#one-hot encoding 된 결과
y = np.array([[1,0],[0,1],[0,1],[1,0]])
x
```

```python
y
```

```python
model = Sequential()
model.add(Dense(units=32, input_dim=2, activation="sigmoid"))
#출력층
model.add(Dense(units=2,#y열의 개수
                activation="softmax"))#다중분류는 활성화 함수가 softmax
```

```python
model.compile(loss="categorical_crossentropy",#다중분류 categorical_crossentropy
             optimizer="RMSprop",
             metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(x,y,epochs=2000)
```

```python
model.evaluate(x,y,verbose=0)
```

```python
#숫자가 더 높은 열이 정답
model.predict(x)
```

```python
model.predict_classes(x)
```
