# 데이터 로드(이미지 파일명만)

```python
import os
```

```python
train_cats_dir = '../data/cats_and_dogs_filtered/train/cats/'
train_dogs_dir = '../data/cats_and_dogs_filtered/train/dogs/'
test_cats_dir = '../data/cats_and_dogs_filtered/test/cats/'
test_dogs_dir = '../data/cats_and_dogs_filtered/test/dogs/'
```

```python
train_cat_fnames = os.listdir(train_cats_dir)
print(train_cat_fnames[:5])
print("개수 :",len(train_cat_fnames),"\n")

train_dog_fnames = os.listdir(train_dogs_dir)
print(train_dog_fnames[:5])
print("개수 :",len(train_dog_fnames),"\n")

test_cat_fnames = os.listdir(test_cats_dir)
print(test_cat_fnames[:5])
print("개수 :",len(test_cat_fnames),"\n")

test_dog_fnames = os.listdir(test_dogs_dir)
print(test_cat_fnames[:5])
print("개수 :",len(test_cat_fnames))
```

# 이미지 2000개 업로드

```python
from matplotlib.pyplot import imread
```

```python
def load_images(default_path, filenames):
    images = []

    for filename in filenames:
        filepath = default_path + filename

        image = imread(filepath) # 파일 읽기
        images.append(image) # 이미지 파일 리스트에 추가

    images = np.array(images) # 이미지 파일 numpy 배열로 변환

    return images
```

```python
X_train_cats = load_images(train_cats_dir, train_cat_fnames)
X_train_dogs = load_images(train_dogs_dir, train_dog_fnames)
print(X_train_cats.shape)
print(X_train_dogs.shape)
```

```python
X_test_cats = load_images(test_cats_dir, test_cat_fnames)
X_test_dogs = load_images(test_dogs_dir, test_dog_fnames)
print(X_test_cats.shape)
print(X_test_dogs.shape)
```

# 이미지 사이즈 조절
- 이미지 크기가 제각각이라 맞춰줘야함
- 사이즈 조절 후 X_train이름으로 합치기

```python
from PIL import Image
```

```python
def resize_image(original_images, size):
    resized_images = []
    
    for image in original_images:
        #resized_image = imresize(image, size) # 1.3.0 버전문제로 없어짐
        resized_image = np.array(Image.fromarray(image).resize(size))
        resized_images.append(resized_image)
        
    #계산 속도를 올리기 위해 array형태로 변환
    resized_images = np.array(resized_images)
    
    return resized_images
```

```python
X_train_cats_224 = resize_image(X_train_cats,(224,224))
print(X_train_cats_224.shape)

X_train_dogs_224 = resize_image(X_train_dogs,(224,224))
print(X_train_dogs_224.shape)
```

```python
X_test_cats_224 = resize_image(X_test_cats,(224,224))
print(X_test_cats_224.shape)

X_test_dogs_224 = resize_image(X_test_dogs,(224,224))
print(X_test_dogs_224.shape)
```

# train, test데이터 고양이 강아지 합치기

```python
X_train = np.concatenate((X_train_cats_224, X_train_dogs_224))
X_train.shape
```

```python
X_test = np.concatenate((X_test_cats_224, X_test_dogs_224))
X_test.shape
```

# label 만들기(고양이 : 0, 강아지 : 1)

```python
y_train = np.array([0]*1000 + [1]*1000)
print(y_train[:10])
print(y_train[-10:])
print(y_train.shape)
```

```python
y_test = np.array([0]*500 + [1]*500)
print(y_test[:10])
print(y_test[-10:])
print(y_test.shape)
```

# CNN 전이학습

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense,Dropout
```

```python
model = Sequential()

model.add(Conv2D(filters=64,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False, # 학습방지
                input_shape = (224, 224, 3)))

model.add(Conv2D(filters=64,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(MaxPooling2D(pool_size = (2,2),
                      strides = (2,2)))

model.add(Conv2D(filters=128,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(Conv2D(filters=128,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(MaxPooling2D(pool_size = (2,2),
                      strides = (2,2)))


model.add(Conv2D(filters=256,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(Conv2D(filters=256,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))


model.add(Conv2D(filters=256,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(MaxPooling2D(pool_size = (2,2),
                      strides = (2,2)))


model.add(Conv2D(filters=512,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))


model.add(Conv2D(filters=512,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(Conv2D(filters=512,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(MaxPooling2D(pool_size = (2,2),
                      strides = (2,2)))


model.add(Conv2D(filters=512,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
               trainable = False))

model.add(Conv2D(filters=512,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(Conv2D(filters=512,
                kernel_size = (3,3),
                activation = 'relu',
                padding = 'same',
                trainable = False))

model.add(MaxPooling2D(pool_size = (2,2),
                      strides = (2,2)))

model.load_weights("../data/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5")

model.add(Flatten())

model.add(Dense(units = 128,
               activation = 'relu'))

model.add(Dense(units = 1,
               activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy',
              optimizer = "adam",
              metrics = ['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint("model_save/cat_dog_{epoch:04d}-{val_accuracy:.4f}.hdf5",
                            mode="max",
                            monitor="val_accuracy",
                            save_best_only=True)
```

```python
model.fit(X_train,
          y_train,
          epochs = 2,
          validation_data=(X_test, y_test),
          callbacks=[checkpoint])
```

```python
model.evaluate(X_test,y_test)
```
