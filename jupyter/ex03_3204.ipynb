```python
from sklearn.preprocessing import StandardScaler

iris_scaled = StandardScaler().fit_transform(iris_Df2)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

pca.fit(iris_scaled)
print(pca.components_)

iris_pca = pca.transform(iris_scaled)
print(iris_pca.shape)
```

```python
iris_pca_Df = pd.DataFrame(iris_pca,columns = ['pca1','pca2'])
iris_pca_Df.head()
```

```python
iris_pca_Df['target']=iris.target
iris_pca_Df.head()
```

```python
print(pca.explained_variance_ratio_)
```

```python
markers=['^','s','o']

for i, markers in enumerate(markers):
    x_axis_data = iris_pca_Df[iris_pca_Df['target']==i]['pca1']
    y_axis_data = iris_pca_Df[iris_pca_Df['target']==i]['pca2']
    plt.scatter(x_axis_data, y_axis_data, marker = markers, label= iris.target_names[i])

plt.legend()
plt.xlabel('pca1')
plt.ylabel('pca2')
plt.grid()
plt.show()
```
