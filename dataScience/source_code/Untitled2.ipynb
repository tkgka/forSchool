```python
t = np.arange(0,10,0.1)
y = np.sin(t)
plt.plot(t,y, color='red',linestyle=':',linewidth='5',marker='*')
```

```python
plt.plot([1,2,3,4,5])
plt.show()
```

```python
plt.figure(figsize=(5,5))
#sin 함수
plt.plot(t,y, color='red',linestyle=':',linewidth='5',marker='*',label =' sin(t)')
#cosin함수
plt.plot(t,np.cos(t), color='blue', linestyle='--', linewidth='5',marker='o', label = 'cos(t)')
plt.grid()
plt.title("싸인함수")
plt.xlabel('x좌표')
plt.ylabel('y좌표')
plt.ylim(-2,2)
plt.text(6,1.5,"빅데이터분석")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('../data/suggestion.csv',encoding ='euc-kr')
df.head()
```

```python
del df['Unnamed: 0']
#df.drop(['Unnamed: 0'],inplace = True, axis = 1)
#df.drop([0], inplace= True ,axis =0)
df.head()
```

```python
df.set_index('아이디',inplace = True)
df.head()
```

```python
plt.bar(df['제안분야'],df['투표수'])
```

```python
plt.barh(df['제안분야'],df['투표수'])
```

```python
df2 = df.groupby(by="제안분야")[['투표수']].sum()
df2
```

```python
df2 = df2.sort_values(by="투표수",ascending=False)
df2
```

```python
df2['투표수'].sort_values().plot(kind='barh')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("제안분야별 투표수")
df2['투표수'].sort_values().plot(kind='barh', color = 'red')
size = 2
for i in range(size):
    plt.text(df2['투표수'][i],size-i,df2['투표수'][i])
plt.grid()
plt.savefig('제안분야.png')
plt.show()
```
